## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### **Chart 1 - Yearly Analysis**

````python
# Chart - 1 visualization code
event_count = data['year'].value_counts().sort_index().values
years = np.sort(data['year'].unique())

# Setting the figure size and resolution
plt.figure(figsize=(14, 6))
plt.rcParams['figure.dpi'] = 140

# Using a dark grid style for the plot
sns.set_style("darkgrid")

# Plotting the scatter plot
sns.scatterplot(x=years, y=event_count, s=30, color='red', alpha=1, legend=None)

# Plotting the line plot
sns.lineplot(x=years, y=event_count, linewidth=1, alpha=0.7, legend=None, color='red')

# Improving y-axis ticks and labels
plt.yticks(fontsize=10)
plt.xticks(fontsize=10, rotation=45)

# Adding title and axis labels with improved formatting
plt.title('Number of Terrorist Events Per Year', size=14, color='#4f4e4e', weight='bold', pad=20)
plt.xlabel('Year', size=12, color='#4f4e4e', labelpad=15)
plt.ylabel('Count of Terrorist Events', size=12, color='#4f4e4e', labelpad=15)

 #Adding annotations for key years
max_year_idx = np.argmax(event_count)
min_year_idx = np.argmin(event_count)

plt.annotate('Highest', xy=(years[max_year_idx], event_count[max_year_idx]),
              xytext=(years[max_year_idx] + 1, event_count[max_year_idx] + 5),
              arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10, color='black')

plt.annotate('Lowest', xy=(years[min_year_idx], event_count[min_year_idx]),
              xytext=(years[min_year_idx] + 1, event_count[min_year_idx] - 10),
              arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=10, color='black')

# Adjusting layout for better fit
plt.tight_layout()

# Showing plot
plt.show()
````

### 1. **Why did you pick the specific chart?**

Here I chose a line chart with scatter points to visualize the frequency of terrorism events per year due to its ability to effectively display trends over time. This combination allows us to observe the general trend and specific data points, making it easy to identify years with high or low activity. The line chart connects data points chronologically, showcasing how the frequency of events has evolved from 1970 to 2017. Additionally, this chart type facilitates comparative analysis across different periods, helps highlight significant changes, and maintains readability, making it an optimal choice for presenting time-series data.

````python
# Chart - 3 visualization code
# Preparing the data for plotting
coun_terror = data['country'].value_counts()[:15].to_frame()
coun_terror.columns = ['Attacks']
coun_kill = data.groupby('country')['casualties'].sum().to_frame()
coun_terror = coun_terror.merge(coun_kill, left_index=True, right_index=True, how='left')

# Plotting the combined data
ax = coun_terror.plot(kind='bar', width=0.9, figsize=(14, 7), color=['#1f77b4', '#ff7f0e'])

# Adding titles and labels
plt.title('Top 15 Countries by Number of Terrorist Attacks and Casualties', size=16, color='#4f4e4e')
plt.xlabel('Country', size=14, color='#4f4e4e')
plt.ylabel('Count', size=14, color='#4f4e4e')

# Adding legend with customized location
plt.legend(['Number of Attacks', 'Total Casualties'], fontsize=12, loc='upper right')

# Annotating bars with values
for p in ax.patches:
    ax.annotate(f'{int(p.get_height()):,}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                xytext=(0, 10),
                textcoords='offset points',
                fontsize=10, color='black')

# Improving the layout
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
````

````python
# Chart - 4 visualization code
years = np.sort(data['year'].unique())
event_count = data['year'].value_counts().sort_index().values
casualty = data['casualties'].groupby(data['year']).sum().sort_index()

# Setting the figure size and resolution
plt.figure(figsize=(14, 6))
plt.rcParams['figure.dpi'] = 140

# Using a dark grid style for the plot
sns.set_style("darkgrid")

# Plotting the scatter plot for terrorist events
sns.scatterplot(x=years, y=event_count, color='red', alpha=1, legend=None)

# Plotting the scatter plot for casualties
sns.scatterplot(x=casualty.index, y=casualty.values, color='blue', alpha=1, legend=None)

# Plotting the line plot for terrorist events
sns.lineplot(x=years, y=event_count, color='red', alpha=0.6, legend='brief', label='Terrorist Events')

# Plotting the line plot for casualties
sns.lineplot(x=casualty.index, y=casualty.values, color='blue', alpha=0.6, legend='brief', label='Casualties')

# Customizing the y-axis ticks and labels
plt.yticks(fontsize=10)
plt.xticks(fontsize=10, rotation=45)

# Adding title and axis labels with improved formatting
plt.title('Trend of Terrorist Events and Casualties', size=16, color='#4f4e4e', weight='bold', pad=20)
plt.xlabel('Years', size=14, color='#4f4e4e', labelpad=15)
plt.ylabel('Count', size=14, color='#4f4e4e', labelpad=15)

# Adding a legend with customized location
plt.legend(fontsize=12, loc='upper left')

# Highlighting significant trends or peaks with annotations (example provided, modify as needed)
max_events_year = years[np.argmax(event_count)]
max_events_count = np.max(event_count)
plt.annotate('Peak Events', xy=(max_events_year, max_events_count),
             xytext=(max_events_year+2, max_events_count-2000),
             arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=12, color='black')

max_casualties_year = casualty.index[np.argmax(casualty.values)]
max_casualties_count = np.max(casualty.values)
plt.annotate('Peak Casualties', xy=(max_casualties_year, max_casualties_count),
             xytext=(max_casualties_year+2, max_casualties_count-10000),
             arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=12, color='black')

# Adjusting layout for better fit
plt.tight_layout()

# Show plot
plt.show()
````
